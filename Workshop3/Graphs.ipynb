```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y1 = np.cos(x)
plt.title('Signal 1')
plt.plot(x,y1)
```

```python
x = np.linspace(0, 10, 100)
plt.title('Signal 2')
y2 = np.cos(x + 1)
plt.plot(x,y2)
```
